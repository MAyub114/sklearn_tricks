# Confusion Matrix display

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import ExtraTreeClassifier
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt
```

### Create dataset


```python
# Make a binary classification problem
X, y = make_classification(n_samples=200, n_features=5, n_classes=2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

### Use ExtraTreeClass classifier

```python
clf = ExtraTreeClassifier().fit(X_train, y_train)
```

### Plot default confusion matrix

```python
ConfusionMatrixDisplay.from_estimator(clf, X_test, y_test, display_labels=['negative', 'positive'])
plt.show()
```
