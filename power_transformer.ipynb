# Power Transformer

* Many linear models require some transformation of numeric features to make them normally distributed
* `StandardScaler` and `MinMaxScaler` work well for most distributions
* However, these techniques will not work on high skewness data
* Instead, use a logarithmic transformation

```python
from sklearn.preprocessing import PowerTransformer
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load diamonds dataset

* Price and caret features are right skewed
  * Peak is on the left
  * Long tail on the right

```python
diamonds = sns.load_dataset("diamonds")
diamonds[["price", "carat"]].hist(figsize=(10, 5))
plt.show()
```

### Apply Logarithmic transform

```python
# Use Power Transformer (logarithmic transform)
power_transform = PowerTransformer()
diamonds.loc[:, ["price", "carat"]] = power_transform.fit_transform(diamonds[["price", "carat"]])

# Show distribution
diamonds[["price", "carat"]].hist(figsize=(10, 5))
```
