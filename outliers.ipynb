# Outlier detection

Use sklearn EllipticalEnvelope to detect outliers in Gaussian features.

```python
import numpy as np
import seaborn as sns
from sklearn.covariance import EllipticEnvelope
```

```python
# Create a sample normal distribution
X = np.random.normal(loc=5, scale=2, size=50).reshape(-1, 1)

# Fit the estimator
ee = EllipticEnvelope(random_state=0)
_ = ee.fit(X)
```

```python
# Plot histogram with density curve
sns.histplot(X, kde=True, bins=15).set(title="Sample Gaussian distribution (mean=5, std dev=2)")
```

```python
# Test
test = np.array([6, 8, 20, 4, 5, 6, 10, 13]).reshape(-1, 1)
sns.histplot(test, kde=True, bins=15).set(title="Test distribution")
```

```python
# predict returns 1 for an inlier and -1 for an outlier
ee.predict(test)
```

```python
index_locations = np.where(ee.predict(test) == -1)
outliers = test[index_locations]
outliers
```
